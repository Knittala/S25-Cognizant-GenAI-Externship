```python
#recursive function created for factorial
def factorial_recursive(n):
    if n < 0:
        return None
    if n == 0 or n == 1:
        return 1
    else:
        return n * factorial_recursive(n - 1)

#recusive function created for fibonacci
def fibonacci_recursive(n):
    if n < 0:
        return None
    if n <= 1:
        return n
    else:
        return fibonacci_recursive(n-1) + fibonacci_recursive(n-2)

#while statement for the interactive menu element
while True:

#printing the menu options
    print("Recursive Functions Menu:")
    print("1. Calculate the factorial of a number")
    print("2. Find the nth Fibonacci number")
    print("3. Exit")

#taking user input for choice
    choice = input("Enter your choice: ")

#conditional statement if menu option 1 is chosen. Here the user input is taken to find the given number's factorial
    if choice == "1":
        while True:
            number = input("Enter a number to find its factorial: ")
            if number.isdigit() or number.startswith("-") and number[1:].isdigit():
                num = int(number)
                break
            else:
                print("Invalid input. Try again with a valid integer.")

        result = factorial_recursive(num)

#Error statement for occurrences of negative values
        if result is None:
            print("Error: Factorial cannot be computed for negative values.")
#Printing the factorial statement
        else:
            print(f"\nThe factorial of {num} is {result}.")

#conditional statement if menu option 2 is chosen. Here the user input is taken for the position of the Fibonacci number
    elif choice == "2":
        while True:
            index = input("Enter the position of the Fibonacci number: ")
            if index.isdigit() or (index.startswith("-") and index[1:].isdigit()):
                index = int(index)
                break
            else:
                print("Invalid input. Try again with a valid integer.")

        result = fibonacci_recursive(index)

#Error statement for occurrences of negative values
        if result is None:
            print("Error: Fibonacci cannot be computed for negative values.")
#Printing the Fibonacci statement
        else:
            print(f"\nThe {index}th Fibonacci number is {result}.")

#Exiting the program
    elif choice == "3":
        print("Exiting the program... Goodbye!")
        break

#If an invalid menu option is chosen by the user
    else:
        print("Invalid input. Try again with a valid choice from 1 to 3.")
```
