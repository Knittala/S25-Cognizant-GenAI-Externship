```python
#Step 1: Input the Password

#taking user input for the password
password = input("Enter your password: ")
```

```python
#Step 2: Evaluate the Password

#establishing the required length of the password to be greater than or equal to 8 characters
req_length = len(password) >= 8

#initializing score for password strength check
score = 0

upper = False
lower = False
digit = False
special = False

#setting a conditional statement that classifies requirements
for char in password:
    if char.isupper():
        upper = True
    elif char.islower():
        lower = True
    elif char.isdigit():
        digit = True
    elif not char.isalnum():
        special = True

#Determining password strength

#Counting points for score
if req_length:
    score += 1
if upper:
    score += 1
if lower:
    score += 1
if digit:
    score += 1
if special:
    score += 1

#Establishing maximum score and rating system
max_score = 5
rating = ""

if score == max_score:
    rating = "Excellent!"
elif score >= 4:
    rating = "Good"
elif score >= 3:
    rating = "Ok"
elif score >= 2:
    rating = "Weak"


#creating a variable that indicates the password is strong if all requirements are met
strong = req_length and upper and lower and digit and special

#conditional statement that notifies the user if their password is strong. If this is not the case, the user will receive a printed message that indicates what requirement they are missing
if strong:
    print("Your password is strong!")
else:
#Missing all requirements
    if not upper and not lower and not digit and not special:
        print("Your password needs at least one uppercase letter, one lowercase letter, one digit, and one special character.")

#Missing 3 requirements
    elif not upper and not lower and not digit:
        print("Your password needs at least one uppercase letter, one lowercase letter, and one digit.")
    elif not upper and not lower and not special:
        print("Your password needs at least one uppercase letter, one lowercase letter, and one special character.")
    elif not upper and not digit and not special:
        print("Your password needs at least one uppercase letter, one digit, and one special character.")
    elif not lower and not digit and not special:
        print("Your password needs at least one lowercase letter, one digit, and one special character.")
#Missing 2 requirements
    elif not upper and not lower:
        print("Your password needs at least one uppercase letter and one lowercase letter.")
    elif not upper and not digit:
        print("Your password needs at least one uppercase letter and one digit.")
    elif not upper and not special:
        print("Your password needs at least one uppercase letter and one special character.")
    elif not lower and not digit:
        print("Your password needs at least one lowercase letter and one digit.")
    elif not lower and not special:
        print("Your password needs at least one lowercase letter and one special character.")
    elif not digit and not special:
        print("Your password needs at least one digit and one special character.")
#Missing 1 requirement
    elif not upper:
        print("Your password needs at least one uppercase letter.")
    elif not lower:
        print("Your password needs at least one lowercase letter.")
    elif not digit:
        print("Your password needs at least one digit.")
    elif not special:
        print("Your password needs at least one special character.")

#Displaying the password strength to the user
print(f"Your password strength is: {rating}")



```
